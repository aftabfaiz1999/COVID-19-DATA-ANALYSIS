```python
#IMPORTING IMPORTANT LIBRARIES#

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#DATA IMPORTING#

df=pd.read_csv('E:\covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
#TOP 5 RECORDS OF DATA#


df.head()
```

```python
#STATISTICAL INFORMATION OF DATA #

df.describe()
```

```python
#DROPING UNNECESSARY COLUMN #

df.drop(['Latitude','Longitude'],axis=1,inplace=True)
```

```python
#REVIEW OF DATA AFTER DROPING COLUMNS#


df.head()
```

```python
#DEFINEING DATA INTO DF FOR RENAME ITS COLUMNS NAME#


df = df [['Date','State/UT','Confirmed','Deaths','Cured']]
```

```python
#RENAME THE COLUMN#

df.columns=['date','state','confirmed','deaths','cured']
```

```python
#LAST 5 RECORDES#

df.tail()
```

```python
#Max Confirm Cases#
confirmed_cases=df.sort_values(by='confirmed',ascending=False)
confirmed_cases[0:20]
```

```python
#Confirm Cases by States#

plt.rcParams['figure.figsize']=(20,15)
sns.barplot(x='state',y='confirmed',data=confirmed_cases)
plt.xticks(rotation=90)
plt.show()
```

SO CAN OBSERVE THE DATA AND SEE THE MOST CONFIRMED  CASES ARE COMING FROM STATES
Maharashtra,
Delhi,
Tamil Nadu,
Rajasthan,

```python
#Max Deaths by States#
max_deaths=df.sort_values(by='deaths',ascending=False)
max_deaths[0:20]
```

```python
#Deaths by States#

plt.rcParams['figure.figsize']=(20,8)
sns.barplot(x='state',y='deaths',data=max_deaths,)
plt.xticks(rotation=90)
plt.show()
```

WE CAN SEE MOST DEATHS OCCUR IN ,Maharashtra,Madhya Pradesh,Delhi,Gujarat,Telengana
