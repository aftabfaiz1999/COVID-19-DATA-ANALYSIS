```python
#IMPORTING IMPORTANT LIBRARIES#

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#DATA IMPORTING#

df=pd.read_csv('E:\covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
#TOP 5 RECORDS OF DATA#


df.head()
```

```python
#STATISTICAL INFORMATION OF DATA #

df.describe()
```

```python
#DROPING UNNECESSARY COLUMN #

df.drop(['Latitude','Longitude'],axis=1,inplace=True)
```

```python
#REVIEW OF DATA AFTER DROPING COLUMNS#


df.head()
```

```python
#DEFINEING DATA INTO DF FOR RENAME ITS COLUMNS NAME#


df = df [['Date','State/UT','Confirmed','Deaths','Cured']]
```

```python
#RENAME THE COLUMN#

df.columns=['date','state','confirmed','deaths','cured']
```

```python
#LAST 5 RECORDES#

df.tail()
```

```python
#Max Confirm Cases_state_top10#
confirmed_cases=df.sort_values(by='confirmed',ascending=False)
confirmed_cases_top10_state=confirmed_cases[0:60]
```

```python
#Confirm Cases by States_top10#

plt.rcParams['figure.figsize']=(20,15)
sns.barplot(x='state',y='confirmed',data=confirmed_cases_top10_state)
sns.lineplot(x='state',y='confirmed',data=confirmed_cases_top10_state)
plt.xticks(rotation=45)
plt.show()
```

SO CAN OBSERVE THE DATA AND SEE THE MOST CONFIRMED  CASES ARE COMING FROM STATES
Maharashtra,
Delhi,
Tamil Nadu,
Rajasthan,
Madhya Pradesh


```python
#Max Deaths by States_top10#
max_deaths=df.sort_values(by='deaths',ascending=False)
max_deaths_state_top10=max_deaths.sort_values(by='deaths',ascending=False)
max_deaths_state_top10=max_deaths_state_top10[0:65]
```

```python
#Deaths by States_top10#

plt.rcParams['figure.figsize']=(20,8)
sns.barplot(x='state',y='deaths',data=max_deaths_state_top10)
sns.lineplot(x='state',y='deaths',data=max_deaths_state_top10)
plt.xticks(rotation=45)
plt.show()
```

WE CAN SEE MOST DEATHS OCCUR IN ,Maharashtra,Madhya Pradesh,Delhi,Gujarat,Telengana
